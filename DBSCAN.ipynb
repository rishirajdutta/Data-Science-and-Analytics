# Implementation of clusters of data points using DBSCAN Algorithm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 10, 10
```

```python
from sklearn.cluster import DBSCAN
```

```python
# This will give us a good resolution for our visualization
%config InlineBackend.figure_format = 'retina' 
```

```python
from sklearn.datasets import make_blobs, make_moons, make_circles
```

```python
X, y = make_blobs(n_samples = 300, n_features = 2, centers = 3, cluster_std = 1.0, shuffle = False, random_state = 42) 
```

```python
df = pd.DataFrame(X)
df.columns = ["x", "y"]
```

## Displaying the raw data on a scatter plot

```python
sns.lmplot("x", "y", data = df, fit_reg = False, size = 7)
plt.show()
```

```python
dbscan = DBSCAN(eps = 1,  min_samples = 20)
```

```python
dbscan.fit(X)
```

```python
df["clusters"] = dbscan.labels_
```

## Displaying the data after applying DBSCAN to create optimal clusters

```python
sns.lmplot("x", "y", data = df, fit_reg = False, hue = "clusters", size = 7)
plt.show()
```

### The blue dots with value -1 are the outliers

```python
A, b = make_moons(n_samples = 300, shuffle = False, noise = 0.05, random_state = None) 
```

```python
df = pd.DataFrame(A)
df.columns = ["a", "b"]
```

## Displayin the raw data in a moon shape on a scatter plot

```python
sns.lmplot("a", "b", data = df, fit_reg = False, size = 7)
plt.show()
```

```python
dbscan = DBSCAN(eps = 0.25,  min_samples = 20)
```

```python
dbscan.fit(A)
```

```python
df["clusters"] = dbscan.labels_
```

## Displaying the clusters on the moon shaped data using DBSCAN

```python
sns.lmplot("a", "b", data = df, fit_reg = False, hue = "clusters", size = 7)
plt.show()
```
